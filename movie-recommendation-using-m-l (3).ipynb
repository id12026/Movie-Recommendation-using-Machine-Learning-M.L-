```python
#Functions to load dataset..............

def getMovie(users,items):
    movies = pd.read_csv('/kaggle/input/movies-csv/movies (1).csv').values
    dataset = np.array([a[:3]  for a in movies if(a[1]<items)])
    data = np.zeros((users,items))
    for i in range(len(dataset)):    
        data[int(dataset[i,0]-1),int(dataset[i,1])] = dataset[i,2]
    return data;
```

```python
#Function to find mean...........

def mean(data):
    mean = np.mean(data[(data > 0)])
    if(math.isnan(mean)):
        return 0
    return mean
```

```python
# Function to find user Mean..............  

def userMean(data,users):
    u_mean = np.zeros(users)
    for i in range(users):
        mean = np.mean(data[i,(data[i] > 0)])
        if(not math.isnan(mean)):
            u_mean[i] = mean       
            
    return u_mean
```

```python
# Function to find item Mean..............    

def itemMean(data,items):
    i_mean  = np.zeros(items)
    for i in range(items):
        mean = np.mean( data[data[:,i]>0,i])
        if(not math.isnan(mean)):
            i_mean[i] = mean        
            
    return i_mean
```

```python
def item_ten(data,uMean):
    items = len(data[0])
    iTen = np.zeros(items)
    for i in range(items):
        temp = np.where(data[:,i] !=0)[0]
        if(len(temp) != 0):
            iTen[i] = sum(data[temp,i] - uMean[temp]) / len(temp)
            
        
    return iTen
```

```python
def user_ten(data,iMean):
    users = len(data)
    uTen = np.zeros((users,1))
    for i in range(users):
        temp = np.where(data[i,:] !=0)[0]
        if(len(temp) != 0):
            uTen[i] = sum(data[i,temp] - iMean[temp]) / len(temp)
    return uTen

    
def user_count(data, users):
    u_count  = np.zeros(users)
    for i in range(users):
        u_count[i] = np.count_nonzero(data[i])
    return u_count

def i_count(data, items):
    i_count  = np.zeros(items)
    for i in range(items):
        i_count[i] = np.count_nonzero(data[:,i])
    return i_count


def II_count(data,items):
    itemCount = np.zeros((items,items))
    for i in range(items):
        for j in range(items):
            if i > j : 
                #print(itemCount[j,i])
                itemCount[i,j] = itemCount[j,i]
            else :
                #print(len(np.intersect1d(np.nonzero(data[:,i]), np.nonzero(data[:,j]))))
                itemCount[i,j] = len(np.intersect1d(np.nonzero(data[:,i]), np.nonzero(data[:,j])))
                
    return itemCount

#def like_dev(data,items):  


    
    
def getDataManual():
    user = int(input("Enter no of users : "))
    item = int(input("Enter no of items : "))
    data = np.zeros((user,item))
    
    for i in range(user):
        print("Enter data of row ",i," seperated by ',' : \n")
        ar = input()
        temp = ar.split(',')
        for j in range(item):
            data[i,j] = int(temp[j])
    return data


def getData(data):
    if(data[0] == '1'):
        return pd.read_csv("DATASET/" + data[2:]).values
    if(data == 'ml.csv'):
        return getMovie(1000,1000)[:500,:500]
    mov = pd.read_csv("DATASET/" + data).values    
    data = np.zeros((6040,3952))
    for i in range(len(mov)):
        x=  mov[i][0].split("::")
        a = int(x[0],10)-1
        b = int(x[1],10)-1
        r = int(x[2],10)  
        data[a,b] = r
        
    print("Before Deleting rows and colls :  ")
    count = 0
    for i in range(len(data)):            
         if( np.count_nonzero(data[i])< 30):               
            count += 1            
    print("No of rows less then 50 ratiings : ",count)
        
    count = 0
    for i in range(len(data[0])):            
        if( np.count_nonzero(data[:,i])< 50):               
            count += 1            
    print("No of colls less then 30 ratiings : ",count)
    print("Shape of data matrix : ", data.shape)
   
       
    return preProcess(data)

class data:
        
    
    def deleteDataRandom(self,data):
        d = 7
        users = len(data)
        delItems = np.zeros((users,d))
        actual = np.zeros((users,d))
        for i in range(users):
            temp = np.where(data[i] != 0)[0]  
            #if(len(temp) > (30 + d)):
            ran = random.sample(set(temp),d)
            actual[i] = data[i,ran]
            data[i,ran] = 0
            delItems[i] = ran
        return data,delItems,actual
    
    
    
def preProcess(data):
    
    rth = 30
    cth = 50  
    
    colls = np.where(np.count_nonzero(data,0) < cth)[0]
    print("\ncolls deleting : ",len(colls))
    data = np.delete(data,colls,1)
        
    rows = np.where(np.count_nonzero(data,1) < rth)[0]  
    print("rows deleting : ",len(rows))
    data = np.delete(data,rows,0)   
   
    
    print("\nAfter Deletion of rows and collms : ")
    count = 0
    for i in range(len(data)):            
         if( np.count_nonzero(data[i])< rth):               
            count += 1            
    print("No of rows less then ",rth," ratings : ",count)
        
    count = 0
    print("collums less then 50 are : ")
    for i in range(len(data[0])):  
        temp = np.count_nonzero(data[:,i])
        if( temp < cth):
            print(temp)
            count += 1            
    print("No of colls less then ",cth," ratings : ",count)
    print("Shape of data matrix : ", data.shape)
    
    
    
    return data
```

Scikit-learn is a library in Python that provides many unsupervised and supervised learning algorithms. 
CountVectorizer is a tool used for converting a collection of text documents into a matrix of token counts.
cv = CountVectorizer(max_features=6000, stop_words='english'): This line instantiates a CountVectorizer object named cv. It sets the maximum number of features (words or tokens) to be considered to 6000 using the max_features parameter. It also specifies that common English words (stop words) should be ignored during the tokenization process using the stop_words parameter set to 'english'.

In summary, this line of code initializes a CountVectorizer object with specific settings for feature selection and stop words, enabling the conversion of text data into numerical vectors for further analysis or modeling.

```python
#CountVectorizer counts text(TAGS) converted into vectors.
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 6000, stop_words = 'english')
```

```python
#Stem words
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stemmer(x):
    stem_word = []
    string = " "
    
    for i in x.split():
        stem_word.append(ps.stem(i))
        
    return string.join(stem_word)
```

```python
final_data['tags'] = final_data['tags'].map(lambda x: stemmer(x))
```

```python
vectors = cv.fit_transform(final_data['tags']).toarray()
```

Cosine similarity:-

Cosine similarity in the recommendation system is used with the same principle of cosine angles, where even if the similarity of the content is less similar it would be considered as the least recommended content, and for higher similarity of contents, the recommendations generated would be at the top. Cosine similarity is also used in textual data to find the similarity between the vectorized texts from the original text document.

```python
#Cosine Similarity measures similarity between two vectors.
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
similarity[1]
```

```python
def recommend(movie):
    movie_index = final_data[final_data['movie_title'] == movie].index[0]
    distances = similarity[movie_index]
    recommended_movies_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x: x[1])[1:9]
    
    for i in recommended_movies_list:
        print(final_data.iloc[i[0]].movie_title)
```

```python
recommend('spider-man: no way home')
```
